```python
import time

def get_data_for_df(end_time,api_key,end_time_switch):
    if end_time_switch:
        matches_list = get_matches_ids(99,end_time,None,api_key)
    else:
        matches_list = get_matches_ids(99,None,end_time,api_key)
    data=[]
    for match in matches_list:
        match_url = f"https://europe.api.riotgames.com/lol/match/v5/matches/{match}?api_key={api_key}"
        match_get = requests.get(match_url)
        match_info = match_get.json()
        my_index = match_info["metadata"]["participants"].index("ZpfBYkSvCESi5fHTUqogFrChHgadmwoPjA6EnHdWiPTHP5CpkT7KGDqSrLwAGKA6ePUNXLWtOH2MQw")
        game_info = match_info["info"]["participants"][my_index]
        data.append({
                "date": time.strftime('%Y-%m-%d %H:%M:%S', time.gmtime(match_info["info"]["gameCreation"]/1000)),
                "game_id": game_info["championName"],
                "role": game_info["teamPosition"],
                "kda": game_info["challenges"]["kda"],
                "skillshotsHit": game_info["challenges"]["skillshotsHit"],
                "skillshotsDodged": game_info["challenges"]["skillshotsDodged"],
                "laneMinionsFirst10Minutes": game_info["challenges"]["laneMinionsFirst10Minutes"],
                "damage": game_info["challenges"]["damagePerMinute"],
                "total_minions_killed": game_info["totalMinionsKilled"],
                "visionScore": game_info["challenges"]["visionScorePerMinute"],
                "goldPerMinute": game_info["challenges"]["goldPerMinute"],
                "time_played_minutes": game_info["timePlayed"]/60,
                "win": game_info["win"]
        })    
    return data
```

   - Popis: Maxim√°ln√≠ hloubka strom≈Ø v Random Forest. Tento parametr omezuje maxim√°ln√≠ poƒçet √∫rovn√≠ v ka≈æd√©m rozhodovac√≠m stromu.
   - Hodnoty: `range(2, 24, 2)` (tj. 2, 4, 6, ..., 22)
   - 

2. **`criterion`**:
   - Popis: Funkce pro mƒõ≈ôen√≠ kvality dƒõlen√≠ v jednotliv√Ωch uzlech stromu. 
   - Hodnoty: `['entropy', 'gini']`


3. **`n_estimators`**:
   - Popis: Poƒçet strom≈Ø v lese (Random Forest). Vy≈°≈°√≠ poƒçet strom≈Ø zlep≈°uje v√Ωkon modelu, ale zvy≈°uje v√Ωpoƒçetn√≠ n√°roƒçnost.
   - Hodnoty: `range(3, 40, 5)` (tj. 3, 8, 13, ..., 38)


#### **Ztr√°tov√° funkce**
Pro vyhodnocen√≠ √∫spƒõ≈°nosti modelu jsme zvolili **p≈ôesnost na validaƒçn√≠ mno≈æinƒõ** (*accuracy*). Tato metrika je vhodn√°, proto≈æe na≈°e data jsou vyv√°≈æen√°, a tud√≠≈æ nen√≠ pot≈ôeba pou≈æ√≠vat pokroƒçilej≈°√≠ metriky, jako je F1-score nebo ROC-AUC. 

#### **D≈Øvod volby metriky**
- P≈ôesnost je jednoduch√° na interpretaci a poskytuje p≈ôehledn√Ω obraz o celkov√©m v√Ωkonu modelu.
- Pokud by data byla nevyv√°≈æen√°, volili bychom jinou metriku, jako nap≈ô√≠klad F1-score nebo ROC-AUC.

#### **Shrnut√≠**
Optimalizoval jsem kombinace parametr≈Ø **`max_depth`**, **`criterion`** a **`n_estimators`** pro nalezen√≠ nejlep≈°√≠ho modelu pomoc√≠ validaƒçn√≠ mno≈æiny na z√°kladƒõ metriky p≈ôesnosti.

Z√°rove≈à jsem zvolil dva modely a jedin√Ω rozd√≠l mezi nimi je absence p≈ô√≠znaku **`kda`**. Jeliko≈æ oƒçek√°v√°m, kv≈Øli vysok√© korelaci **`kda`** s **`win`**, jej√≠ d≈Øle≈æitost.


**Pozn√°mky**

Zkou≈°el jsem v√≠ce hyperparametr≈Ø jako nap≈ô√≠klad **`max_features`**, **`min_samples_split`**,**`min_samples_leaf=1`**, kter√© nevedly k drastick√©mu zlep≈°en√≠ a tr√©novan√≠ trvalo kv≈Øli tomu p≈ô√≠li≈° dlouho, tud√≠≈æ jsem se rozhodl je nepou≈æ√≠t.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import ParameterGrid
import sklearn.metrics as metrics
# hyperparametry
param_grid = {
    'max_depth': range(2,24,2), 
    'criterion': ['entropy', 'gini'],
    'n_estimators': range(3,40,5)
}
#-----------------------------With KDA---------------------------------------------------
param_comb = ParameterGrid(param_grid)
val_acc1 = []
train_acc1 = []
for params in param_comb:
    clf = RandomForestClassifier(**params,n_jobs=-1,random_state=123)
    clf.fit(Xtrain1, ytrain1)
    train_acc1.append(metrics.accuracy_score(ytrain1, clf.predict(Xtrain1)))
    val_acc1.append(metrics.accuracy_score(yval1, clf.predict(Xval1)))

#-----------------------------Without KDA---------------------------------------------------
val_acc2 = []
train_acc2 = []
for params in param_comb:
    clf = RandomForestClassifier(**params,n_jobs=-1,random_state=123)
    clf.fit(Xtrain2, ytrain2)
    train_acc2.append(metrics.accuracy_score(ytrain2, clf.predict(Xtrain2)))
    val_acc2.append(metrics.accuracy_score(yval2, clf.predict(Xval2)))
```
